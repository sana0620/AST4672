### Create two 3D arrays (float 64), one for the objects and one for the darks. Your arrays should have sizes (nobj, ny, nx) and (ndark, ny, nx) respectively. Print the shapes of these arrays.




```python
obj_array  = np.zeros( ( nobj, ny , nx ) )
dark_array = np.zeros( ( ndark, ny , nx ) )
```

```python
print( obj_array.shape , dark_array.shape )
```

### Populate your data cubes! Read the target and dark data into the arrays. Save the last header from each set in a variable (e.g., objhead and darkhead). Read and print the date of observation (DATE-OBS) from each of these.

```python
for o in range( nobj ):
    with fits.open(os.path.join(datadir, objfile)) as hdul:
    #star_file = datadir + objfile[ o ] + fext
        star_file = hdul[0].data
    if o < nobj-1:
        obj_array[ o, :, : ] = fits.getdata( star_file )
    else:
        obj_array[ o, :, : ], objhead = fits.getdata( star_file, header= True )
    
```

```python
for d in range( ndark ):
    dark_file = datadir + darkfile[ d ] + fext
    
    if d < ndark-1:
        dark_array[ d, :, : ] = fits.getdata( dark_file )
    else:
        dark_array[ d, :, : ], darkhead = fits.getdata( dark_file, header=True )
```
